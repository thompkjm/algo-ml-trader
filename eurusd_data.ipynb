```python
import pandas as pd
from pathlib import Path
import numpy as np
from sklearn.cluster import KMeans
```

```python
eur_usd = pd.read_csv(
    Path('../algo-ml-trader/resources/6mo of 30min eur usd.csv'),
    index_col='Date',
    parse_dates=True,
    infer_datetime_format=True
)
eur_usd['slope'] = np.nan
eur_usd.head()
```

```python
eur_usd['slope'] = eur_usd['Last Price'].diff()/eur_usd['Date'].diff()
eur_usd.head()
```

```python
eur_usd = eur_usd.reset_index()
eur_usd
```

```python
eur_usd['10_sma']=eur_usd["Last Price"].rolling(10).mean()
eur_usd['20_sma']=eur_usd["Last Price"].rolling(20).mean()
```

```python
eur_usd.head()
```

```python
eur_usd["signal"]=np.where(eur_usd["10_sma"]>eur_usd["20_sma"],1,0)
```

```python
eur_usd['slope'].describe()
```

```python
eur_usd['Entry/Exit'] = eur_usd['signal'].diff()
eur_usd.head(40)
```
